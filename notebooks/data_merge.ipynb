```python
import pandas as pd

steamcharts = pd.read_csv("../data/raw/steamcharts.csv")
steam = pd.read_csv("../data/raw/steam.csv")
```

 **COMBINE STEAM CHART DATA ABOUT APP WITH ITS METADATA FROM steam dataset** 





```python
print("SteamCharts columns:")
print(steamcharts.columns)

print("\nSteam metadata columns:")

print(steam.columns)
```

```python
steam = steam.rename(columns={"appid": "steam_appid"})
```

```python
merged = steamcharts.merge(
    steam,
    on="steam_appid",
    how="left"
    
)

keep_cols = [
    # SteamCharts
    'month', 'avg_players', 'gain', 'gain_percent', 'peak_players',
    'name_x',  'steam_appid',
    # Steam metadata
    'release_date', 'genres', 'steamspy_tags',
    'positive_ratings', 'negative_ratings',
    'average_playtime', 'median_playtime',
    'owners', 'price', 'name_y'
]

merged = merged[keep_cols].copy()
```

**EXPORT NEW DATASET TO steam_merged_csv**

```python

print(merged.head())
merged["genres"].isna().sum()

# exclude name_y since it is the same as name_x
merged = merged.drop(columns=["name_y"])

merged.to_csv("../data/processed/steam_merged.csv", index=False)
```

```python
import pandas as pd

steamcharts = pd.read_csv("../data/raw/steamcharts.csv")
steam = pd.read_csv("../data/raw/steam.csv")
```

```python
print("SteamCharts columns:")
print(steamcharts.columns)

print("\nSteam metadata columns:")

print(steam.columns)
```

```python
steam = steam.rename(columns={"appid": "steam_appid"})
```

```python
merged = steamcharts.merge(
    steam[["steam_appid", "genres", "name", "steamspy_tags"]],
    on="steam_appid",
    how="left"
    
)
```

```python

print(merged.head())
merged["genres"].isna().sum()

merged.to_csv("../data/processed/steam_merged.csv", index=False)
```
